## Step1:  Data Prepration

```python
import pandas as pd
import os
import shutil
```

#### Create the data for positive samples(Github)

```python
#Create the data for positive samples(Github)
FILE_PATH="covid-chestxray-dataset-master/metadata.csv"
IMAGES_PATH="covid-chestxray-dataset-master/images"
```

```python
df=pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df
```

```python
TARGET_DIR="Dataset/Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print("Covid Folder created")
```

```python
cnt=0
for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row["view"]=="PA":
        filename=row["filename"]
        image_path=os.path.join(IMAGES_PATH,filename)
        image_copy_path=os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving image",cnt)    
        cnt +=1
print(cnt)   
```

#### Create the data for Normal Samples(Kaggle)

```python
#Sampling of IMages from Kaggle
import random
KAGGLE_FILE_PATH ="chest_xray/train/NORMAL"
TARGET_NORMAL_DIR="Dataset/Normal"
```

```python
images_names=os.listdir(KAGGLE_FILE_PATH)
```

```python
random.shuffle(images_names)
```

```python
for i in range(196):
    image_name = images_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("Coping image",i)
```
